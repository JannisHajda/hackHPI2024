```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import pathlib as path
```

```python
directory = path.Path('../data/loadprofiles/summer')
json_files = directory.glob('*.json')

data = {}
for json_file in json_files:
    with open(json_file, 'r') as f:
        name = path.Path(json_file).stem
        data[name] = json.load(f)
```

```python
for key, values in data.items():
    pd.Series(values['Electro']).plot(label=key)

plt.title('Summer Electricity')
plt.legend()
```

```python
for key, values in data.items():
    if 'Heating' in values:
        pd.Series(values['Heating']).plot(label=key)
    else:
        pd.Series([0 for i in range(24)]).plot(label=key)

plt.title('Summer Heating')
plt.legend()
```

```python
pd.Series(winter_data['RESIDENTIAL']['Heating']).plot(label='Heating')
pd.Series(winter_data['RESIDENTIAL']['Electro']).plot(label='Electro')
plt.title('Winter Consumption for Housing')
plt.xlabel('Hour')
plt.ylabel('kW/m²')
plt.legend()
```
